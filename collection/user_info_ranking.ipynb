```python
import pandas as pd
from urllib import request
import requests
import json
import time
import os
```

```python
def get_user_space(mid):
    # 获取用户信息(空间)

    # url
    url = 'https://api.bilibili.com/x/space/acc/info?mid=%s'%(mid)
    # 请求网页
    response = request.urlopen(url)
    # 读取内容
    content = response.read()
    # 读取json
    dirt = json.loads(content)
    # 写入用户信息(空间)
    user_space = dirt['data']
    # 返回用户信息(空间)
    return user_space
```

```python
def mkdir(path):
    # 判断路径是否存在
    isExists=os.path.exists(path)

    if not isExists:
        # 如果不存在则创建目录
        os.makedirs(path) 
        return True
    else:
        return False
```

```python
def user_info_output(mid):
    # 用户信息输出

    # 获取用户信息(空间)
    user_space = get_user_space(mid)
    
    # 判断数据获取情况
    if (user_space):
        # 创建DataFrame
        # 用户信息(空间)
        output_user_space = pd.DataFrame([user_space])

        # 获取当前目录
        current_path= os.getcwd()
        # 创建目录
        mkdir(current_path+'/data/user/{}'.format(mid))

        # 写入
        # 用户信息(空间)
        output_user_space.to_csv(current_path+'/data/user/{}/{}.csv'.format(mid, 'user_space'))
        
        # 完成提示
        print(mid, ': 写入完成')
    else:
        # 失败提示
        print(mid, ': 写入失败')
```

```python
def get_user_id_list():
    # 获取user_id列表

    # 获取当前目录
    current_path= os.getcwd()
    # 读取CSV文件
    user_id = pd.read_csv(current_path+'/preprocess/ranking.csv')
    # 写入user_id列表
    user_id_list = pd.DataFrame(data=user_id, columns=['mid'])
    user_id_list = user_id_list.dropna()
    
    return user_id_list
```

```python
def user_list_to_user_info():
    # 获取用户信息

    # 获取当前目录
    current_path= os.getcwd()
    
    # 获取用户id列表
    user_id_list = get_user_id_list()
    # 获取用户id数量
    user_id_list_num = user_id_list.shape[0]
    # 循环
    for i in range(0, user_id_list_num):
        # 获取用户id
        mid = int(user_id_list.iat[i, 0])
        # 判断路径是否存在
        isExists=os.path.exists(current_path+'/data/user/{}'.format(mid))
        if not isExists:
            try:
                # 获取用户信息
                user_info_output(mid)
                # 空行
                print('\n')
                # 等待2秒
                time.sleep(2)
            except Exception as e:
                # 获取失败
                print(mid, ': 获取失败')
                print('错误: ', e, '\n')
                # 等待2秒
                time.sleep(2)
    # 全部完成提示
    print('用户信息写入完成')
```

```python
user_list_to_user_info()
```
