```python
import pandas as pd
from urllib import request
import requests
import simplejson as json
import time
import os
```

```python
def get_video_detail(bvid, sessdata):
    # 获取视频详细信息

    # url
    url = 'https://api.bilibili.com/x/web-interface/view?bvid=%s'%(bvid)
    # Cookies
    cookies = {'SESSDATA': sessdata}
    # 请求网页
    response = requests.get(url, cookies=cookies)
    # 读取内容
    content = response.text
    # 读取json
    dirt = json.loads(content)
    # 写入视频详细信息
    video_detail = dirt['data']
    # 返回视频详细信息
    return video_detail
```

```python
def get_video_tag(bvid):
    # 获取视频TAG

    # url
    url = 'https://api.bilibili.com/x/tag/archive/tags?bvid=%s'%(bvid)
    # 请求网页
    response = request.urlopen(url)
    # 读取内容
    content = response.read()
    # 读取json
    dirt = json.loads(content)
    # 写入视频TAG
    video_tag = dirt['data']
    # 返回视频TAG
    return video_tag
```

```python
def bvID_to_CID(bvid):
    # bvID转CID(oid) 

    # url
    url = 'https://api.bilibili.com/x/player/pagelist?bvid=%s' %bvid
    # 请求网页
    response = request.urlopen(url)
    # 读取内容
    content = response.read()
    # 读取json
    dirt = json.loads(content)
    # 获取数量
    num = len(dirt['data'])
    # 创建列表
    cid_list = []
    # 循环
    for i in range(0, num):
        # 写入CID到列表
        cid_list.append(dirt['data'][i]['cid'])
    # 返回列表
    return cid_list
```

```python
def mkdir(path):
    # 判断路径是否存在
    isExists=os.path.exists(path)

    if not isExists:
        # 如果不存在则创建目录
        os.makedirs(path) 
        return True
    else:
        return False
```

```python
def video_info_output(bvid, sessdata, cate_id):
    # 视频信息输出

    # 获取oid列表
    oid_list = bvID_to_CID(bvid)
    # 获取视频数量
    num = len(oid_list)

    # 获取视频详细信息
    video_detail = get_video_detail(bvid, sessdata)
    # 获取视频TAG
    video_tag = get_video_tag(bvid)
    
    # 判断数据获取情况
    if (video_detail and video_tag):
        # 创建DataFrame
        # 视频详细信息
        output_video_detail = pd.DataFrame([video_detail])
        # 视频TAG
        output_video_tag = pd.DataFrame(data=video_tag)

        # 获取当前目录
        current_path= os.getcwd()
        # 创建目录
        mkdir(current_path+'/data/video/{}/{}'.format(cate_id, bvid))

        # 写入
        # 视频详细信息
        output_video_detail.to_csv(current_path+'/data/video/{}/{}/{}.csv'.format(cate_id, bvid, 'video_detail'))
        # 视频TAG
        output_video_tag.to_csv(current_path+'/data/video/{}/{}/{}.csv'.format(cate_id, bvid, 'video_tag'))

        # 完成提示
        print(bvid, ': 写入完成')
    else:
        # 失败提示
        print(bvid, ': 写入失败')
```

```python
def ranking_to_video_info(sessdata, cate_id):
    # 读取排行榜并获取视频信息

    # 获取当前目录
    current_path= os.getcwd()
    # 读取CSV文件
    ranking = pd.read_csv(current_path+'/data/ranking/{}.csv'.format(cate_id))
    # 获取bvid列表
    bvid_list = pd.DataFrame(data=ranking, columns=['bvid'])
    # 获取bvid数量
    bvid_num = bvid_list.shape[0]

    # 循环
    for i in range(0, bvid_num):
        # 获取bvid
        bvid = bvid_list.iat[i, 0]
        # 判断路径是否存在
        isExists=os.path.exists(current_path+'/data/video/{}/{}'.format(cate_id, bvid))
        if not isExists:
            try:
                # 获取视频信息
                video_info_output(bvid, sessdata, cate_id)
                # 等待5秒
                time.sleep(5)
                # 空行
                print('\n')
            except Exception as e: 
                # 获取失败
                print(bvid, ': 获取失败')
                print('错误: ', e, '\n')
                # 等待5秒
                time.sleep(5)
    # 全部完成提示
    print(cate_id, ': 视频信息写入完成')
```
