```python
import pandas as pd
from urllib import request
import requests
import simplejson as json
import time
import os
```

```python
def bvid_to_avid(bvid):
    table = 'fZodR9XQDSUm21yCkr6zBqiveYah8bt4xsWpHnJE7jL5VG3guMTKNPAwcF' #码表
    tr = {} #反查码表
    #初始化反查码表
    for i in range(58):
        tr[table[i]] = i
    s = [11, 10, 3, 8, 4, 6] #位置编码表
    xor = 177451812 #固定异或值
    add = 8728348608 #固定加法值

    def bv2av(x):
        r = 0
        for i in range(6):
            r += tr[x[s[i]]] * 58 ** i
        return (r - add) ^ xor

    return bv2av(bvid)
```

```python
def mkdir(path):
    # 判断路径是否存在
    isExists=os.path.exists(path)

    if not isExists:
        # 如果不存在则创建目录
        os.makedirs(path) 
        return True
    else:
        return False
```

```python
def get_comment(avid):
    # 获取评论
    
    # 创建DataFrame
    comment = pd.DataFrame()
    # 循环
    for page in range (1, 5):
        # url
        url = 'https://api.bilibili.com/x/v2/reply?type=1&oid=%s&sort=1&ps=49&pn=%s'%(avid, page)
        # 请求网页
        response = request.urlopen(url)
        # 读取内容
        content = response.read()
        # 读取json
        dirt = json.loads(content)
        # 写入评论
        comment = comment.append(dirt['data']['replies'])
        # 等待5秒
        time.sleep(5)
    # 返回评论
    return comment
```

```python
def ranking_to_comment(cate_id):
    # 读取排行榜并获取评论
    
    # 获取当前目录
    current_path= os.getcwd()
    # 读取CSV文件
    ranking = pd.read_csv(current_path+'/data/ranking/{}.csv'.format(cate_id))
    # 获取bvid列表
    bvid_list = pd.DataFrame(data=ranking, columns=['bvid'])
    # 获取bvid数量
    bvid_num = bvid_list.shape[0]
    
    # 判断路径是否存在
    isExists=os.path.exists(current_path+'/data/comment/{}'.format(cate_id))
    if not isExists:
        # 创建目录
        mkdir(current_path+'/data/comment/{}'.format(cate_id))
    
    # 循环
    for i in range(0, bvid_num):
        # 获取bvid
        bvid = bvid_list.iat[i, 0]
        # 判断路径是否存在
        isExists=os.path.exists(current_path+'/data/comment/{}/{}.csv'.format(cate_id, bvid))
        if not isExists:
            try:
                # 获取avid
                avid = bvid_to_avid(bvid)
                # 获取评论列表
                comment_list = pd.DataFrame(data=get_comment(avid))
                # 写入评论列表
                comment_list.to_csv(current_path+'/data/comment/{}/{}.csv'.format(cate_id, bvid))
                # 完成提示
                print(cate_id, '-', bvid, ': 写入完成')
                # 空行
                print('\n')
            except: 
                # 获取失败
                print(cate_id, '-', bvid, ': 获取失败', '\n')
    # 全部完成提示
    print(cate_id, ': 评论写入完成')
```

```python
# 分区号
ranking_to_comment('146')
```
