```python
import pandas as pd
import os
import numpy as np
```

```python
def get_file_list():
    # 获取当前目录
    current_path= os.getcwd()
    # 文件路径
    path = current_path+'/data/ranking'
    # 获取文件列表
    filelist = os.listdir(path)
    # 返回文件列表
    return filelist
```

```python
# 获取当前目录
current_path= os.getcwd()
# 获取文件列表
filelist = get_file_list()
# 获取文件数量
file_num = len(filelist)
# 创建DataFrame
ranking = pd.DataFrame()

# 循环
for i in range(0, file_num):
    # 读取ranking
    temp = pd.read_csv(current_path+'/data/ranking/{}'.format(filelist[i]))
    # 删除第一列
    temp = temp.drop(columns=['Unnamed: 0'])
    # 标记分区号
    temp['video_zone'] = filelist[i].split('.')[0]
    # 写入ranking
    ranking = ranking.append(temp)
    # 完成提示
    print(filelist[i], ': 处理完成', '\n')

# 写入ranking到文件
ranking.to_csv(current_path+'/preprocess/ranking.csv', index=False)
# 完成提示
print('写入完成', '\n')
```
